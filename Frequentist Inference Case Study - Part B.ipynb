__Q2:__ The administrator is concerned that the actual average charge has fallen below 12,000, threatening the hospital's operational model. On the assumption that these data represent a random sample of charges, how would you justify that these data allow you to answer that question? And what would be the most appropriate frequentist test, of the ones discussed so far, to apply?

__A:__ We rely on the CLT to justify that these data will allow us to answer the question. The CLT states that for a sufficiently large sample size, the sampling distribution of the mean will be approximately normal, even if they are highly skewed, as seen here. We will perform a one-sample t-test to evaluate whether the true population mean is likely below 12,000. 

__Q3:__ Given the nature of the administrator's concern, what is the appropriate confidence interval in this case? A ***one-sided*** or ***two-sided*** interval? (Refresh your understanding of this concept on p. 399 of the *AoS*). Calculate the critical value and the relevant 95% confidence interval for the mean, and comment on whether the administrator should be concerned.

__A:__ We will calculate the 95% confidence interval in this case and use a one-sided interval, because we are only interested in whether the medical charges has fallen below 12,000. 95% One-Sided Lower Bound: 12725.49. Based on this sample, we are 95% confident that the true average medical charge is at least 12725.49. Since the average has not fallen below the 12000, the administrator should not be concerned just yet. 

```python
n = len(medical.charges)
s_mean = medical.charges.mean()
s_std = medical.charges.std()
confidence_level = 0.95
standard_error = s_std / math.sqrt(n)
df = n -1
t_critical = t.ppf(1 - confidence_level, df)

lower_bound = s_mean + (t_critical * standard_error)

print(t_critical)
print(f"95% One-Sided Lower Bound: {lower_bound:.2f}")
```

The administrator then wants to know whether people with insurance really are charged a different amount to those without.

__Q4:__ State the null and alternative hypothesis here. Use the _t_-test for the difference between means, where the pooled standard deviation of the two groups is given by:
\begin{equation}
s_p = \sqrt{\frac{(n_0 - 1)s^2_0 + (n_1 - 1)s^2_1}{n_0 + n_1 - 2}}
\end{equation}

and the *t*-test statistic is then given by:

\begin{equation}
t = \frac{\bar{x}_0 - \bar{x}_1}{s_p \sqrt{1/n_0 + 1/n_1}}.
\end{equation}

(If you need some reminding of the general definition of ***t-statistic***, check out the definition on p. 404 of *AoS*). 

What assumption about the variances of the two groups are we making here?

__A:__ 
* Null Hypothesis(Ho): There is no significant difference between the mean medical charges for patients with insurance and patients without insurance. Alternative Hypothesis (H1): There is a statistically significant difference between the mean medical charges for patients with insurance and patients without insurance.
* We are making an assumption that both broups have equal variances. 

__Q5:__ Perform this hypothesis test both manually, using the above formulae, and then using the appropriate function from [scipy.stats](https://docs.scipy.org/doc/scipy/reference/stats.html#statistical-tests) (hint, you're looking for a function to perform a _t_-test on two independent samples). For the manual approach, calculate the value of the test statistic and then its probability (the p-value). Verify you get the same results from both.

__A:__ T-Stat: 11.8933, P-Value: 0.0000e+00. This is my result in both the manual and scipy.stats tests. With P-value essentially zero, we reject the null hypothesis

```python
group_insured = medical[medical['insuranceclaim'] == 1]['charges']
group_uninsured = medical[medical['insuranceclaim'] == 0]['charges']

ni, nu = len(group_insured), len(group_uninsured)
si, su = group_insured.std(), group_uninsured.std()
mi, mu = group_insured.mean(), group_uninsured.mean()

# Pooled Standard Deviation
sp = np.sqrt(((ni - 1) * si**2 + (nu - 1) * su**2) / (ni + nu - 2))

# T-Statistic
t_stat = (mi - mu) / (sp * np.sqrt(1/ni + 1/nu))

# Degrees of Freedom
dfp = ni + nu - 2

# P-Value (Two-Sided)
p_value = 2 * (1 - t.cdf(abs(t_stat), dfp))

print(f"Pooled T-Stat: {t_stat:.4f}, P-Value: {p_value:.4e}")
```

```python
# Perform Welch's T-Test (Two-Sided)
# We use two-sided because the administrator wants to know if they are "different"
t_stat_w, p_value_w = stats.ttest_ind(group_insured, group_uninsured, equal_var=True)

print(f"Insured Mean:   {group_insured.mean():.2f}")
print(f"Uninsured Mean: {group_uninsured.mean():.2f}")
print(f"T-Statistic:    {t_stat_w:.4f}")
print(f"P-Value:        {p_value_w:.4e}") # Using scientific notation for very small p-values
```
